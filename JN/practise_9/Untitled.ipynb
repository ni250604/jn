```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("train.csv")
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
```

```python
df["Sex"] = df["Sex"].apply(fun1)
```
