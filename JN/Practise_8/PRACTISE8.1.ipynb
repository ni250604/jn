```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("train.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data["Age"]=data["Age"].fillna(data["Age"].mean)
```

```python
data.isna().sum()
```

```python
def fun1(value):
    if(value=='male'):
        return 1
    else:
        return 0
```

```python
data["Sex"].value_counts()
```

```python
data["Embarked"].value_counts()
```

```python
data["Sex"]=data["Sex"].apply(fun1)
```

```python
data["Sex"].value_counts()
```

```python
def fun2(value):
    if(value=='S'):
        return 0
    elif(value=='C'):
        return 1
    else:
        return 2
```

```python
data["Embarked"]=data["Embarked"].apply(fun2)
```

```python
data["Embarked"].value_counts()
```

```python
data=data.drop("Cabin",axis=1)
numeric_counts=data.select_dtypes(include=['float64','int64'])
sns.heatmap(numeric_counts.corr(),annot=True)
```
